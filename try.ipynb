```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):
  def __init__(self, in_features=4, h1=5, h2=6, out_f=3):
    super().__init__()
    self.fc1 = nn.Linear(in_features, h1)
    self.fc2 = nn.Linear(h1, h2)
    self.fc3 = nn.Linear(h2, out_f)

  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)

    return x
```

```python
torch.manual_seed(41)

model = Model()
```
